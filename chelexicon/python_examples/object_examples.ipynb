```python
class ObjectExample:
    """
    This class provides examples of the attributes and methods available natively in python.
    
    .. py:class:: ObjectExample
        .. py:function:: __init__()
    """

example_object = ObjectExample()
```

```python
def dynamically_added_method(self) -> None:
    """
    This is
    :py:function:: dynamically_added_method
    :param self:
    :return:
    """
    print("Wow, we added a new method to the class dynamically!")

setattr(ObjectExample, "dynamically_added_method", dynamically_added_method)
```

```python
def __init__(self) -> None:
    """
    The __init__ method is called when the object is created. Use this method to declare any class attributes that will be used by the object.
    """
    print("Entering __init__.")
        
setattr(ObjectExample, "__init__", __init__)
example_object = ObjectExample()
```

```python
def __enter__(self) -> None:
    """
    The __enter__ method is called when the object is used with a context manager. This is the method that makes your class compatible with the 'with' statement.
    """
    print("Entering __enter__.")
    
def __exit__(self, exc_ty, exc_val, tb) -> None:
    """
    The __exit__ method is called when the object exits the context manager. You can use this method to clean up after you object when its used with a context manager. Example: requests.Session objects execute session.close() in their __exit__ method to clean up after the connection is no longer needed.
    """
    print("Entering __exit__.")
    
setattr(ObjectExample, "__enter__", __enter__)
setattr(ObjectExample, "__exit__", __exit__)

def with_example() -> None:
    with example_object as example:
        print("Entering 'with' block.")
        print("Look! I'm inside a with block! ^~^")
        print("Exiting 'with' block.")
    
with_example()
```
